```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import gensim
from gensim.models import Word2Vec
from joblib import dump
```

```python
import numpy as np
import pandas as pd
import string
import re
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def load_data():
    data = pd.read_csv('cleaned_data.csv', names=['Review','Sentiment','Sentence_Removal','Lower_Case','Sen_Out_StopWord','stem'])
    return data
```

```python
roman_urdu_df= load_data()
roman_urdu_df.head()
```

```python
roman_urdu_df = roman_urdu_df.drop(roman_urdu_df.index[0])
roman_urdu_df = roman_urdu_df.reset_index(drop=True)
roman_urdu_df.head()
```

```python
roman_urdu_df.isnull().sum()
roman_urdu_df.dropna(inplace = True)
```

```python
from sklearn.model_selection import train_test_split

train_features,test_features, train_target, test_target = train_test_split(roman_urdu_df['stem'], roman_urdu_df['Sentiment'],
                                                          random_state=42,
                                                          test_size=0.2)
```

```python
count_vectorizer = CountVectorizer(lowercase=False, stop_words=None, ngram_range=(1, 3))
train_features_cv = count_vectorizer.fit_transform(train_features)
test_features_cv= count_vectorizer.fit_transform(test_features)
train_target_cv=count_vectorizer.fit_transform(train_target)
test_target_cv=count_vectorizer.fit_transform(test_target)
dump(train_features_cv, 'train_features_cv.joblib')
dump(test_features_cv, 'test_features_cv.joblib')
dump(train_target_cv, 'train_target_cv.joblib')
dump(test_target_cv, 'test_target_cv.joblib')
print(train_features_cv.shape)
```
